```python
states = ['Michigan', 'California', 'Texas', 'Indiana']
#your code goes here
new_df = house[house['state'].isin(states)]

new_df['date'] = pd.to_datetime(new_df['date'])

new_df.head()
```

b) Use the FacetGrid from seaborn to create line plots to represent house_price_index for each state. Please make the line color red, set ticks on x axis for years ['1980', '2000', '2020'], which you may have to convert date type again, put those years as labels, and set the titles for each small plot to their state names. In addition, set the y label to 'house price index' and remove the x label since it's obvious that these are years.

```python
# prompt: Use the FacetGrid from seaborn to create line plots to represent house_price_index for each state.
# Please make the line color red, set ticks on x axis for years ['1980', '2000', '2020'],
# convert date type again, put those years as labels, and set the titles for each small plot to their state names.
# In addition, set the y label to 'house price index' and remove the x label since it's obvious that these are years

import pandas as pd
import matplotlib.pyplot as plt

# Convert 'date' column to datetime objects if not already done
new_df['date'] = pd.to_datetime(new_df['date'])

# Create the FacetGrid
g = sns.FacetGrid(new_df, col="state", col_wrap=4, height=4)

# Map the line plot function
g.map(plt.plot, "date", "house_price_index", color="red")

# Set x-axis ticks and labels using FacetGrid.axes
for ax in g.axes.flat:
    ax.set_xticks([pd.to_datetime(year) for year in ['1980', '2000', '2020']])
    ax.set_xticklabels(['1980', '2000', '2020'])

# Set titles for each subplot
g.set_titles("{col_name}")

# Set y-axis label
g.set_ylabels("house price index")

# Remove x-axis label
g.set_xlabels("")


plt.show()
```

```python
# prompt: Draw a lineplot for 'house price perc', but color them based on the 'state' name which will have four
#lines in a single plot. Set a legend which will be at the bottom of your graph and have them horizantally
#positioned next to eachother. Again, set the y label to house_price_perc and remove the x label

import matplotlib.pyplot as plt
# Assuming 'new_df' from the previous code block is available

plt.figure(figsize=(12, 6))  # Adjust figure size as needed

for state in states:
    subset = new_df[new_df['state'] == state]
    plt.plot(subset['date'], subset['house_price_perc'], label=state)

plt.ylabel('house_price_perc')
plt.xlabel('') # Removing x label as requested.

plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.2), ncol=4, fancybox=True, shadow=True)
plt.show()
```
